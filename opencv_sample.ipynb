```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# 1.画像の読み込みと表示
- cv2.imread()
- plt.imshow()

jupyter で画像を表示する際はmatplotlibのimshowを使う
cv2.imshowだと別のウィンドウになりあまりおいしくない

```python
#画像の読み込み
img = cv2.imread("sample.jpg")
```

```python
#画像の表示
#そのままだと色が変になってしまう
plt.imshow(img)
```

```python
#opencv　BGR(blue, green, red)
#matplotlib RGB
#三色の順番のルールが違うのでRGBに変換してあげる必要がある→めんどくさい

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# 2.リサイズ

```python
img.shape
```

```python
w = 200
h = 100
img_resize = cv2.resize(img,(w,h))

#　表示
img_rgb = cv2.cvtColor(img_resize, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#　縦横を元の画像の半分にする
size = img.shape[:2]
h = size[0]//2
w = size[1]//2

img_resize = cv2.resize(img,(w,h))

#　表示
img_rgb = cv2.cvtColor(img_resize, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```
