# 0.準備編

グラフがかっこよくなるのでseabornを使っています。

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# グラフがかっこよくなるおまじない 
import seaborn as sns
# plot style default
sns.set()
# plot style setting : small  <<  "paper" < "notebook" < "talk" < "poster" << bigger
sns.set_context("talk")
```

## サンプルデータ
今回は適当な乱数を使う。
- x : ランダム乱数100個
- y : ランダム乱数100個

```python
x = np.random.randn(100)
x
```

```python
y = np.random.randn(100)
y
```

# 1.グラフ書く編

## 折れ線

```python
# ふつう
plt.plot(x)
```

```python
# 重ねる
plt.plot(x)
plt.plot(y)
```

## 棒グラフ

```python
# 普通
plt.bar(range(100), x)
```

```python
# 重ねる
plt.bar(range(100), x)
plt.bar(range(100), y)
```

## 散布図

```python
# ふつう
plt.scatter(x,y)
```

## ヒストグラム

```python
# ふつう
plt.hist(x)
```

```python
# 細かさ設定 大きいと細かくなる
plt.hist(x, bins=30)
```

```python
# 重ねる
plt.hist(x)
plt.hist(y)
```

```python
# 重ねて半透明
plt.hist(x, alpha=0.5)
plt.hist(y, alpha=0.5)
```

```python
# 積む
plt.hist([x,y], stacked=True)
```

# 2.デコレーション編

## タイトル

```python
plt.plot(x)
plt.title("Hello Matplotlib")
```

## 軸

```python
# ヨコ軸
plt.plot(x)
plt.xlabel("sample nummber")
```

```python
# タテ軸も
plt.plot(x)
plt.xlabel("sample nummber")
plt.ylabel("value")
```

## 判例

```python
# きほん
plt.plot(x, label='x')
plt.legend()
```

```python
# 複数
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.legend()
```

## 描画する範囲

```python
# 何も指定しない
plt.plot(x)
```

```python
# 0 - 20 のみ
plt.plot(x)
plt.xlim(0,20)
```

## 色の変更

```python
# 赤　'r'
plt.plot(x, c='r')
```

```python
# 緑　'g'
plt.plot(x, c='g')
```

## 複数並べて表示

```python
# 複数表示
plt.plot(x)
plt.show()
plt.hist(x)
plt.show()
```
